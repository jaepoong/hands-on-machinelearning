## 소프트벡터 머신 SVM
* 클래스 사이의 가장 폭(마진)이 넓은 도로를 만드는 것과 같다. 
  * > 도로 경계의 샘플을 서포트 벡터라고 한다.   
    > 마진 오류를 가진 모델을 소프트 마진 분류라 하며 이는 하이퍼파라미터 C로 조정이 가능하다.
![image](https://user-images.githubusercontent.com/62787572/130427747-c5ea7aef-1efe-454f-af4f-bf2103d58ae6.png).  
* 비선형 SVM분류
  * > SVM은 비선형 형태로 분류가 가능하다.
![image](https://user-images.githubusercontent.com/62787572/130430566-7379b9bf-97d9-4582-b4ba-c2cd25054f6a.png)
* 다항식 커널
  * > SVM은 낮은 차수는 잘 표현하지 못하고 높은 차원은 매우 느려진다.
    > kernel tricks 수학적 방식으로 이를 컨트롤이 가능하다.
    > kernel(poly,rbf,linear),degree(차수),coef0(차수의 영향도),C(모델의 이상치 인정값),gamma(각 데이터가 영향을 미치는 범위, 작을수록 커진다.) 등의 하이퍼파라미터로 설정 가능하다. 
![image](https://user-images.githubusercontent.com/62787572/130431670-a4d4427d-8d6b-4bd3-905e-ecea509c8ba3.png)

---
> 위의 gamma의 경우 rbf커널의 유사도함수의 범위를 설정한다.
> 이를통해 해당 범위가 미치는 영역을 판단한다(크기와 영역은 반비례) 위 그림과 같이 gamma가 클 경우 랜드마크의 영향범위가 줄어든다.
>> * 과대적합일 경우 C와 gamma를 감소시키고 과소적합은 그 역이다.
>> * 선형 SVM은 O(N)이고 kernelSVM은 O(N^2~N^3)이다. 따라서 데이터가 클 경우 매우 느려진다.
>>   * 실제 Mnist를 돌려도 30분이 넘게 걸린다.
* SVM 회귀
> SVM을 회귀로 사용 할 경우 마진 내에 최대한 많은 샘플이 들어가게 훈련을 시킨다.

> 위 그림과 같이 마진을 가능한 크게하는 선형의 W와 b를 찾는 것이 svm.C의 목표이다.
---
### 목적함수
> * 결정함수의 기울기는 ||W||(노름)과 같다. 이를 2로 나누면 마진을 2배로 크게하는 것과 같으므로 가중치의 노름을 최소화하는 것이 SVM.C의 목표이다.
> * 미분을 용이하기 위해 ||W||^2 와 같은 W(T)W를 목적함수 즉 손실함수로 사용한다.
> * 여기에 C를 사용해 샘플이 마진을 얼마나 위반할지를 정한다.
>  * 손실함수와 C는 각각 마진을 최대화하고 마진오류를 최소화하는 Trade Of의 관계를 가진다.

6. RBF SVM이 과소적합되면 gamma와 C를 줄여야하나 늘여야하나?
> * gamma는 각 샘플이 영향을 미칠 수 있는 범위이다(반비례). gamma가 커지면 범위가 줄어들고 gamma가 줄어들면 범위가 커진다. 과소적합이 되면 gamma를 증가시켜서 모델의 변별능력을 증가시켜야 한다.
> * C는 모델의 오류에 대한 하이퍼 파라미터이다. C가 늘어나면 모델이 마진 오류를 더 많이 무시하게 된다. 따라서 과소적합이 되면 C를 증가시켜야한다.

9. MNist데이터셋에 SVM분류기를 훈련시켜라.


```python
# 데이터셋 준비
from sklearn.datasets import fetch_openml
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
mnist = fetch_openml('mnist_784', version=1, cache=True, as_frame=False)

x = mnist["data"]
y = mnist["target"].astype(np.uint8)

x_train = x[:5000].reshape(-1,784)
y_train = y[:5000]
x_test = x[5000:6000].reshape(-1,784)
y_test = y[5000:6000]
scaler=StandardScaler()
x_train,x_test=scaler.fit_transform(x_train),scaler.fit_transform(x_test)
```

```python
# OvR을 사용해서 사이킷런

from sklearn.multiclass import OneVsRestClassifier
clf = OneVsRestClassifier(SVC(kernel='rbf',C=5,gamma=0.5))
clf.fit(x_train, y_train)
```

```python
# 모델 저장하기
from sklearn.externals import joblib
joblib.dump(clf, 'filename.pkl') 
```

```python
#모델 불러오기
clf= joblib.load('filename.pkl') 
```

```python
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier
y_pred = clf(x_test)
accuracy_score(y_test,y_pred)
```
