### 성능 측정
  *  교차검증 
    * cross_val
    * StratifiedKFold 는 클래스별 비율이 유지되도록 폴드 만듬.
      * for문으로 교차검증 시행
      * 교차검증과 비슷하지만 세부 조정이 필요하면 하는 방식.
  * 불균형 데이터셋의 경우 정확도를 측정 지표로 선호하지 않는다.
  * 오차행렬
    * 분류기 성능 평가 지표.
    * confusion_matrix(y_train,y_train_pred)로 오차행렬 생성.
      * TN,FP,FN,TP순. 정밀도 : 양성중 실제 양성 비율, 재현율 : 분류기가 정확하게 감지한 양성 샘플의 비율.
      * 정밀도 : precision(y_train,y_train_pred)
      * 재현율 : recall_score(y_train,y_train_pred)
      * F1스코어 : 정밀도와 재현율의 조화 평균.
      * 정밀도 재현율 트레이드 오프 존재.
      * clf.decision_function으로 점수 확인해 입계값 지정 가능.
      * ROC곡선 : 거짓 양성비율에 대한 진짜 양성비율의 ㅈ곡선
        * fpr,tpr,thresholds = roc_curve(y_train,y_scores)로 확인 가능.
        * AUC 곡선 아래 멵적으로 분류기 비교 가능

### 에러 분석

  * 이 행렬의 주대각선을 np.fill_diagonal(~,0)으로 그래프 그리면 무엇에 오류가 나는지 확인 가능하다.
  * 위 과정 이후 오류가 많은 부분을 수정할 수 있다.
